```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new directed graph
G = nx.DiGraph()

# Add nodes and edges for the first graph
G.add_edge('X', 'Y', label='c')

# Create a second graph
H = nx.DiGraph()

# Add nodes and edges for the second graph
H.add_edge('X', 'M', label='a')
H.add_edge('M', 'Y', label='b')
H.add_edge('X', 'Y', label="c'")

# Position the nodes manually for the first graph
pos_G = {'X': (0, 0), 'Y': (1, 0)}

# Position the nodes manually for the second graph
pos_H = {'X': (0, 0), 'M': (0.5, 1.5), 'Y': (1, 0)}

# Draw the first graph
plt.figure(figsize=(8, 8))

plt.subplot(211)
plt.gca().set_xlim(-0.2,1.2)
plt.gca().set_ylim(-0.4,1.5)
nx.draw(G, pos_G, with_labels=True, node_size=6000, node_color='white', edgecolors='black', font_weight='bold',font_size=18)
nx.draw_networkx_edge_labels(G, pos_G, edge_labels={('X', 'Y'): 'c'}, font_color='black', font_size=18)

# Draw the second graph
plt.subplot(212)
plt.gca().set_xlim(-0.2,1.2)
plt.gca().set_ylim(-0.7,2.2)
nx.draw(H, pos_H, with_labels=True, node_size=6000, node_color='white', edgecolors='black', font_weight='bold', font_size=18)
nx.draw_networkx_edge_labels(H, pos_H, edge_labels={('X', 'M'): 'a', ('M', 'Y'): 'b', ('X', 'Y'): "c'"}, font_color='black', font_size=18)


plt.show()
```
